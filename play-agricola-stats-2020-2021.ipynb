## Check PWR and ADP for all play-agricola.com games in 2020-2021

Here we print out for all cards that was found. If a card was changed during the period, then there will be different versions of it in the list

```python
file="./Data/2020_2021_pwr.pkl"
df=open_df(file)
card_names=list(set(df["card_id"].tolist()))
pwr=make_pwr_df(df,card_names)
```

We print only the PWR of cards which are drafted at least 50 times. 

```python
print_pwr(pwr,threshold=50)
```
